```python
import re
import os
import numpy as np
```

```python
def readLines(file):
    ''' Function accept the file and return its all lines'''
    with open(file) as fh:
        return fh.readlines()
    
def wordBag(lines):
    wbag = {}
    for line in lines:
        for word in line.split():
            wbag[word]=wbag.get(word,0)+1
    return wbag
        
# reading the mbox data
with open("data/mbox.txt") as fh:
    print(len(fh.readlines()))
```

```python
print(readLines("data/mbox.txt")[:3])
print()
print(wordBag(readLines("data/mbox.txt")[:10]))
```
